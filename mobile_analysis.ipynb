### Удаление пустых столбцов

```python
def get_numerical_summary(df):
    total = df.shape[0]
    missing_columns = [col for col in df.columns if df[col].isnull().sum() > 0]
    missing_percent = {}
    nu_cou = {}
    for col in missing_columns:
        null_count = df[col].isnull().sum()
        per = (null_count/total) * 100
        missing_percent[col] = per
        nu_cou[col] = len(df) - null_count
        print("{} : {} ({}%)".format(col, null_count, round(per, 3)))
    return (nu_cou, missing_percent)
```

```python
print("Количество NaN в столбцах")
null_count, missing_percent = get_numerical_summary(df)
```

```python
print("Количество до удаления:",len(df.columns))
for i in null_count.keys():
    if missing_percent[i] > 40:
        df.drop(columns = i, inplace = True)
        
print("Количество после удаления:",len(df.columns))
```

```python
print("Количество NaN в столбцах")
null_count, missing_percent = get_numerical_summary(df)
```

```python
def convert_col(col,unit,med):
    
    na_value = "0" + unit
    df[col].fillna(na_value, inplace=True)
    df.loc[:,col] = df[col].apply(lambda x: float(x.split(unit)[0]))
    
    if med:
        df.loc[(df[col] == 0), col] = df[col].median()
```

```python
convert_col("Кол-воточекнадюйм","ppi",1)
convert_col("Основнаякамера","мп",0)
convert_col("Частотапроцессора","мгц",1)
convert_col("Емкостьаккумулятора","ма*ч",1)
convert_col("Вес","г",1)
convert_col("Длина","мм",1)
convert_col("Ширина","мм",1)
convert_col("Толщина","мм",1)
df["Кол-воядерпроцессора"].fillna(df["Кол-воядерпроцессора"].median(), inplace=True)
df["ВерсияBluetooth"].fillna(0, inplace=True)
df.loc[(df["ВерсияBluetooth"] != 0), "ВерсияBluetooth"] = 1
```
